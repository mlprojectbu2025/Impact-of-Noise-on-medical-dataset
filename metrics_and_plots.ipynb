```python
import numpy as np
import matplotlib.pyplot as plt

def plot_metric_grouped_bar(dataset_name, results, metric, colors):
    all_noise_levels = [0, 5, 10, 15, 20, 25, 30, 35, 40]
    selected_indices = [0, 2, 4, 6, 8]  # Indices of 0, 10, 20, 30, 40
    noise_levels = [all_noise_levels[i] for i in selected_indices]

    models = list(results.keys())
    x = np.arange(len(noise_levels))
    w = 1 / (len(models) + 1)

    plt.figure(figsize=(10, 6), dpi=120)
    for i, model in enumerate(models):
        # Filter only selected metric values
        values = [results[model][metric][j] for j in selected_indices]
        plt.bar(x + i*w, values, w, label=model,
               color=colors.get(model, "#000000"))
    plt.xticks(x + w*(len(models)-1)/2, noise_levels)
    plt.xlabel("Noise (%)")
    plt.ylabel(metric)
    plt.title(f"{dataset_name} - {metric}")
    plt.ylim(0, 1)
    plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
    plt.tight_layout()
    plt.show()

for dataset_name, results in datasets.items():
    for metric in metrics:
        plot_metric_grouped_bar(dataset_name, results, metric, colors)
```
